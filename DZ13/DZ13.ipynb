ЗАДАНИЯ С БАЗОЙ ДАННЫХ:
1. Вытащите данные и найдите средний возраст клиентов из Минска
2. Выведите список всех клиентов с балансом больше 1000
3. Подсчитайте количество транзакций по каждому типу
4. Определите клиентов с наибольшими расходами (отрицательные суммы транзакций)
5. Топ-3 клиента по количеству транзакций
6. Выявление подозрительных операций, операции > n% от баланса


```python
import pandas as pd
import sqlite3

with sqlite3.connect('database.db') as conn:
    query = "SELECT * FROM clients"
    df_clients = pd.read_sql(query, conn)
    query1 = "SELECT * FROM balance"
    df_balance = pd.read_sql(query1, conn)
    query2 = "SELECT * FROM transactions"
    df_transactions = pd.read_sql(query2, conn)
    


    print(f'Средний возраст клиентов из Минска: {df_clients["age"][df_clients["city"] == "Минск"].mean()}\n'
        f'Список клиентов с балансом >1000: \n {df_clients.loc[(df_clients["client_id"] == df_balance["client_id"]) & (df_balance["amount"] > 1000)]}\n'
        f'Количество транзакций deposit: {df_transactions["type"][df_transactions["type"] == "deposit"].count()}\n'
        f'Количество транзакций withdrawal: {df_transactions["type"][df_transactions["type"] == "withdrawal"].count()}\n '
            )
```

ЗАДАНИЯ С API:
1. Количество капсул каждого типа
2. Среднее время между первым и последним запуском
3. Капсулы без информации о запуске
4. Сравнение Dragon 1.0 и Dragon 1.1. Среднее кол-во повторного использования, среднее кол-во посадок

```python
import requests
import pandas as pd


response = requests.get('https://api.spacexdata.com/v3/capsules')


data = response.json()
df = pd.DataFrame(data)

print(
    f'Количество капсул каждого типа: {df.groupby(df['type'])['capsule_id'].count()}\n'
    f'Капсулы без информации о запуске: \n{df[df["original_launch"].isnull()]}\n'
)
```
