Задача 4. Анализ возраста (если бы он был)
Задача:
- Добавьте столбец с возрастом (случайные значения от 18 до 65).
- Разделите пользователей на возрастные группы: 18-25, 26-35, 36-50, 51+.
- Подсчитайте количество пользователей в каждой группе.
- Найдите средний возраст для каждой группы.


```python
import pandas as pd
import random as rd

all_info = pd.read_csv("hw_DE_12.csv") # загрузка данных
all_info["Age"] = [rd.randint(18, 65) for i in range(len(all_info))] # добавление возраста

all_info["AgeGroup"] = pd.cut(all_info["Age"], bins=[17, 25, 35, 50, 65], labels=["18–25", "26–35", "36–50", "51+"]) #разделение на группы

group_counts = all_info["AgeGroup"].value_counts()

group_means = all_info.groupby("AgeGroup")["Age"].mean()

#результат
print(f"Количество пользователей по группам: {group_counts}")
print(f"\nСредний возраст по группам: {group_means}")
```
